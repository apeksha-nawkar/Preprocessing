```python
import pandas as pd
import numpy as np

import sklearn
from sklearn.ensemble import IsolationForest
```

```python
df=pd.read_csv("iris(1).csv")
```

```python
df
```

```python
finaldf= pd.get_dummies(df)
```

```python
finaldf
```

```python
model=IsolationForest(random_state=40,contamination=0.2)
model.fit(finaldf)
```

```python
pred=model.predict(finaldf)
pred
```

```python
finaldf['scores']=model.decision_function(finaldf)
```

```python
finaldf['anomaly']=model.predict(finaldf.iloc[:,0:7])
```
